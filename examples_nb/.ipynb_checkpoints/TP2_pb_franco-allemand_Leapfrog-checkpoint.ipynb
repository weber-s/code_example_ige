Ce notebook reprend le précédent en ajoutant une variation temporelle du terme source.

Le problème repris est celui du problème franco-allemand vu en TD (P0), sur les pluies acides. On a formulé le problème de l'évolution de la masse d'un polluant (par exemple le soufre = dioxyde de soufre + sulfate) par un modèle de boîte où l'on décrit la conservation de la masse (voir cours Daniel Jacob, équations 3.7, 3.8) :

\begin{equation}
\frac{dm}{dt} = S - k m
\end{equation}


Pour la résolution numérique, on a exprimé la dérivée temporelle via les valeurs de $m$ à deux temps consécutifs $t_n$ et $t_{n+1}$, $dm/dt = (m_{n+1}-m_n)/dt$, ce qui permet de calculer la masse à $t_{n+1}$ à partir de celle à $t_{n}$ :

$$ m_{n+1} = m_{n} + (S - k m_{n} ) \times dt $$

On obtient donc l'évolution temporelle de la masse de polluant, que l'on peut tracer.

## Importation de modules utiles

```python
import numpy as np                        # pour le calcul numérique
import matplotlib.pyplot as plt           # pour les plots, et dessous, pour que les plots apparaissent dans le notebook
%matplotlib inline
```

## Initialisation de paramètres et de la masse au temps 0

Il faut donner des valeurs aux paramètres qui ne sont pas explicitement obtenus par notre modèle au cours du temps. Et il faut donner une valeur initiale à la masse pour démarrer le calcul de l'évolution temporelle.

On donne des noms un peu explicite aux variables pour bien les reconnaitre.

```python
Source = 200.*1000./365.                    # en T/jour
kphc = 1./4.                                # constante physico-chimique, en /jour
kout = 1e-5*86400.                          # constante de perte par transport (V/d) en /jour
mso2 = 0.                                   # Masse initiale de SO2 (en T)
### Données relative au temps
dt = 0.05                                     # en jour
Nstep = 180                                 # Nombre de pas de temps que l'on prévoit de calculer. Variable entière.
### Terme source avec une variabilité journalière
time_steps = np.arange(Nstep)
Src = Source * (1 + 0.9 *np.sin(2*np.pi*dt*time_steps))
plt.plot(Src)
print("Source mean: "+str(Source))
print("Source max: {max}   Source min: {min}".format(max=Src.max(), min=Src.min()))
```

## Definition du terme de droite de l'équation

Il s'avère qu'on utilise très souvent 

$$
\frac{dm_t}{dt} = S - k m_t
$$

on l'écrit donc sous forme de fonction. Puisque c'est le « terme de droite » (right hand side), on l'appelle `rhs`.

```python
def rhs(m_in, ind):
    return Src[ind] - ktot * m_in
```

# Boucle temporelle

On va calculer les valeurs de $m$ au cours du temps itérativement grâce à 

$$
m_{t+1} = m_t + \frac{dm_t}{dt} \times dt.
$$

Mais auparavant, il faut prévoir objets nécessaires à la sauvegarde des données calculées (ici, seulement un vecteur pour stocker la série temporelle de $m$). Eventuellement, on définit des paramètres qui allège l'écriture des équations (ici : $ktot$).

```python
mso2_series = np.zeros(Nstep)               # Vecteur de Nstep valeurs, mais non rempli
ktot = kout + kphc
mso2 = 0
for t in range(Nstep-1):
    mso2 = mso2 + rhs(mso2, t) * dt
    mso2_series[t+1] = mso2
```

```python
plt.plot(mso2_series, label="Euler direct")
plt.legend()
```

# Une amélioration : le Leap-Frog

L'équation différentielle initiale est toujours 

$$
\frac{dm_t}{dt}= E_t - k\times m_t = f(m_t)
$$

Cette fois, on utilise la dérivée moyenne entre $t+1$ et $t-1$, c'est-à-dire la dérivée à $t$, pour estimer $m_{t+1}$ :

$$
m_{t+1} - m_{t-1} = 2 \Delta t \times f(m_t)
$$

soit,

$$
m_{t+1} = m_{t-1} + 2 \Delta t \times f(m_t)
$$

On utilise ainsi la variation moyenne entre 2 pas de temps : l'endroit où la masse et sa dérivée sont calculées ne sont plus synchronisées, mais se « saute par dessus », d'où le « leap-frog ».


```python
mso2_series_LF = np.zeros(Nstep)               # Vecteur de Nstep valeurs, rempli de 0

mso2 = 0  # on réinitialise la masse à 0

# La masse à t1 est estimé grâce à une schéma d'Euler direct
mso2_series_LF[1] = mso2 + rhs(mso2, 0) * dt
# ... et on commence le leap-frog à t1
for t in range(1,Nstep-1):
    # on est au temps t.
    # on récupère la mase à t-1 et à t
    mso2_b = mso2_series_LF[t-1]
    mso2   = mso2_series_LF[t]
    # on calcul la masse à t+1 grâce à la dérivée à t et m à t-1
    mso2_n = mso2_b + rhs(mso2, t) * 2 * dt
    # et on la garde en mémoire pour t+1
    mso2_series_LF[t+1] = mso2_n
```

```python
plt.plot(mso2_series, label="Euler forward")
plt.plot(mso2_series_LF, label="Leap-frog")
plt.legend()
```

On voit bien que le résultat obtenu est abérant. On met ici en évidence un phénomène étonnant des schémas d'intégration numériques du type leap-frog : il existe 2 « modes » dans la solution. Le mode « physique », qui nous intéresse, et un artefact : le mode numérique. Pour les équations non linéaire, ce mode prend de l'importance au fur et à mesure et devient prépondérant sur la solution physique. Le mode numérique provient du fait que l'on créé des erreures sur les pas de temps pair différents des pas de temps impair, que l'on propage au fur et à mesure.

Il existe 2 solutions pour prévenir ces solutions abéranntes : 

 * intercaler un schéma ne faisant intervenir que 2 pas de temps réguliérement, pour neutraliser l'amplification du mode numérique.
 * filter les résultats avec un filtre du type Robert-Asselin (voir par exemple l'introduction de http://d-scholarship.pitt.edu/28208/1/YongLi_PhD_Thesis_1.pdf). 
 
Le groupe avec Emmanuel Cosme a implémenté cette technique. Alors allons-y nous aussi ! 


```python
mso2 = 0
mso2_series_LFRA = np.zeros(Nstep)               # Vecteur de Nstep valeurs

# première valeur estimée par un schéma d'Euler direct
mso2_series_LFRA[1] = mso2 + rhs(mso2, 0)*dt
for t in range(1,Nstep-1):
    # on est au temps t
    mso2_b = mso2_series_LFRA[t-1]
    mso2   = mso2_series_LFRA[t]
    # Leap-frog
    mso2_n = mso2_b + rhs(mso2, t)*2*dt
    mso2_series_LFRA[t+1] = mso2_n
    # filtre de robert asselin
    mso2_f = mso2 + 0.1*(mso2_b - 2 * mso2 + mso2_n)
    mso2_series_LFRA[t] = mso2_f
```

```python
plt.plot(mso2_series, label="Euler forward")
plt.plot(mso2_series_LFRA, label="Leap-frog+filter")
plt.legend()
```
